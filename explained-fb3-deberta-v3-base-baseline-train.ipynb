# Utils

```python
# ====================================================
# Utils
# ====================================================
def MCRMSE(y_trues, y_preds):
    scores = []
    idxes = y_trues.shape[1]
    for i in range(idxes):
        y_true = y_trues[:,i]
        y_pred = y_preds[:,i]
        score = mean_squared_error(y_true, y_pred, squared=False) # RMSE
        scores.append(score)
    mcrmse_score = np.mean(scores)
    return mcrmse_score, scores

def get_score(y_trues, y_preds):
    mcrmse_score, scores = MCRMSE(y_trues, y_preds)
    return mcrmse_score, scores

def get_logger(filename=OUTPUT_DIR+'train'):
    from logging import getLogger, INFO, StreamHandler, FileHandler, Formatter
    logger = getLogger(__name__)
    logger.setLevel(INFO)
    handler1 = StreamHandler()
    handler1.setFormatter(Formatter("%(message)s"))
    handler2 = FileHandler(filename=f"{filename}.log")
    handler2.setFormatter(Formatter("%(message)s"))
    logger.addHandler(handler1)
    logger.addHandler(handler2)
    return logger

LOGGER = get_logger()

def seed_everything(seed=42):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    
seed_everything(seed=42)
```

# Dataset

```python
# ====================================================
# Define max_len
# ====================================================
lengths = []
tk0 = tqdm(train['full_text'].fillna("").values, total=len(train))
for text in tk0:
    length = len(tokenizer(text, add_special_tokens=False)['input_ids'])
    lengths.append(length)
CFG.max_len = max(lengths) + 3 # cls & sep & sep
LOGGER.info(f"max_len: {CFG.max_len}")
```

```python
# ====================================================
# Dataset
# ====================================================
def prepare_input(cfg, text):
    inputs = cfg.tokenizer.encode_plus(
        text, 
        return_tensors=None, 
        add_special_tokens=True, 
        max_length=CFG.max_len,
        pad_to_max_length=True,
        truncation=True
    )
    for k, v in inputs.items():
        inputs[k] = torch.tensor(v, dtype=torch.long)
    return inputs


class TrainDataset(Dataset):
    def __init__(self, cfg, df):
        self.cfg = cfg
        self.texts = df['full_text'].values
        self.labels = df[cfg.target_cols].values

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, item):
        inputs = prepare_input(self.cfg, self.texts[item])
        label = torch.tensor(self.labels[item], dtype=torch.float)
        return inputs, label
    

def collate(inputs):
    mask_len = int(inputs["attention_mask"].sum(axis=1).max())
    for k, v in inputs.items():
        inputs[k] = inputs[k][:,:mask_len]
    return inputs
```

# Model

```python
# ====================================================
# Model
# ====================================================
class MeanPooling(nn.Module):
    def __init__(self):
        super(MeanPooling, self).__init__()
        
    def forward(self, last_hidden_state, attention_mask):
        input_mask_expanded = attention_mask.unsqueeze(-1).expand(last_hidden_state.size()).float()
        sum_embeddings = torch.sum(last_hidden_state * input_mask_expanded, 1)
        sum_mask = input_mask_expanded.sum(1)
        sum_mask = torch.clamp(sum_mask, min=1e-9)
        mean_embeddings = sum_embeddings / sum_mask
        return mean_embeddings
    

class CustomModel(nn.Module):
    def __init__(self, cfg, config_path=None, pretrained=False):
        super().__init__()
        self.cfg = cfg
        if config_path is None:
            self.config = AutoConfig.from_pretrained(cfg.model, output_hidden_states=True)
            self.config.hidden_dropout = 0.
            self.config.hidden_dropout_prob = 0.
            self.config.attention_dropout = 0.
            self.config.attention_probs_dropout_prob = 0.
            LOGGER.info(self.config)
        else:
            self.config = torch.load(config_path)
        if pretrained:
            self.model = AutoModel.from_pretrained(cfg.model, config=self.config)
        else:
            self.model = AutoModel(self.config)
        if self.cfg.gradient_checkpointing:
            self.model.gradient_checkpointing_enable()
        self.pool = MeanPooling()
        self.fc = nn.Linear(self.config.hidden_size, 6)
        self._init_weights(self.fc)
        
    def _init_weights(self, module):
        if isinstance(module, nn.Linear):
            module.weight.data.normal_(mean=0.0, std=self.config.initializer_range)
            if module.bias is not None:
                module.bias.data.zero_()
        elif isinstance(module, nn.Embedding):
            module.weight.data.normal_(mean=0.0, std=self.config.initializer_range)
            if module.padding_idx is not None:
                module.weight.data[module.padding_idx].zero_()
        elif isinstance(module, nn.LayerNorm):
            module.bias.data.zero_()
            module.weight.data.fill_(1.0)
        
    def feature(self, inputs):
        outputs = self.model(**inputs)
        last_hidden_states = outputs[0]
        feature = self.pool(last_hidden_states, inputs['attention_mask'])
        return feature

    def forward(self, inputs):
        feature = self.feature(inputs)
        output = self.fc(feature)
        return output
```

# Loss

```python
# ====================================================
# Loss
# ====================================================
class RMSELoss(nn.Module):
    def __init__(self, reduction='mean', eps=1e-9):
        super().__init__()
        self.mse = nn.MSELoss(reduction='none')
        self.reduction = reduction
        self.eps = eps

    def forward(self, y_pred, y_true):
        loss = torch.sqrt(self.mse(y_pred, y_true) + self.eps)
        if self.reduction == 'none':
            loss = loss
        elif self.reduction == 'sum':
            loss = loss.sum()
        elif self.reduction == 'mean':
            loss = loss.mean()
        return loss
```

# Helpler functions

```python
# ====================================================
# Helper functions
# ====================================================
class AverageMeter(object):
    """Computes and stores the average and current value"""
    def __init__(self):
        self.reset()

    def reset(self):
        self.val = 0
        self.avg = 0
        self.sum = 0
        self.count = 0

    def update(self, val, n=1):
        self.val = val
        self.sum += val * n
        self.count += n
        self.avg = self.sum / self.count


def asMinutes(s):
    m = math.floor(s / 60)
    s -= m * 60
    return '%dm %ds' % (m, s)


def timeSince(since, percent):
    now = time.time()
    s = now - since
    es = s / (percent)
    rs = es - s
    return '%s (remain %s)' % (asMinutes(s), asMinutes(rs))



# train for 1 epoch. return train loss and train score 
def train_fn(fold, train_loader, model, criterion, optimizer, epoch, scheduler, device):
    
    # start the training mode of model. 
    model.train()
    scaler = torch.cuda.amp.GradScaler(enabled=CFG.apex)

    # losses is average of loss for every sameple on current epoch.
    losses = AverageMeter()
    start = end = time.time()
    global_step = 0


    # step is number of batches.
    # (inputs, labels) is a batch of inputs and labels to this batch.
    for step, (inputs, labels) in enumerate(train_loader):
        inputs = collate(inputs)


        # move inputs and labels to device.
        for k, v in inputs.items():
            inputs[k] = v.to(device)
        labels = labels.to(device)

        # why need to calculate batch_size here b.c. not every batch has the same size. (usually the last batch)
        batch_size = labels.size(0)

        # autocase is used to automatically scale the loss. Scale means choosing the right type of number used to store loss: float16 or float32.
        # Enables autocasting for the forward pass (model + loss)
        with torch.cuda.amp.autocast(enabled=CFG.apex):
            y_preds = model(inputs)
            # loss is a tensor with shape (1,).
            loss = criterion(y_preds, labels)


        # This is great idea to use gradient accumulation. 
        
        if CFG.gradient_accumulation_steps > 1:
            loss = loss / CFG.gradient_accumulation_steps

        # update loss of this batch to total loss. # loss is actually a scalar value. 
        losses.update(loss.item(), batch_size)

        # backward loss to calculate gradient. # we should understand the below line as loss.backward(). Scaler.scale(loss) is a independent value.
        scaler.scale(loss).backward()

        # clip_grad_norm_ is used to prevent gradient explosion.
        normedGrad = torch.nn.utils.clip_grad_norm_(model.parameters(), CFG.max_grad_norm)
        if (step + 1) % CFG.gradient_accumulation_steps == 0:

            # update model parameters. # optimzer should understand as trinh toi uu hoa. 
            # https://discuss.pytorch.org/t/what-does-the-backward-function-do/9944/2
            scaler.step(optimizer)

            # Update scaler factor.
            scaler.update()

            # x.grad is saved in somewhere after backward.
            optimizer.zero_grad()
            global_step += 1


            # update learning rate when update model parameters.
            if CFG.batch_scheduler:
                scheduler.step()

                
        end = time.time()
        if step % CFG.print_freq == 0 or step == (len(train_loader)-1):
            print('Epoch: [{0}][{1}/{2}] '
                  'Elapsed {remain:s} '
                  'Loss: {loss.val:.4f}({loss.avg:.4f}) '
                  'Grad: {normedGrad:.4f}  '
                  'LR: {lr:.8f}  '
                  .format(epoch+1, step, len(train_loader), 
                          remain=timeSince(start, float(step+1)/len(train_loader)),
                          loss=losses,
                          normedGrad=normedGrad,
                          lr=scheduler.get_lr()[0]))

        if CFG.wandb:
            # loss and learning rate are logged every step => In wandb, you can see the learning rate and loss's graph have a lot of value.
            # avg_val_loss, for e.g, has only 4 values. 
            wandb.log({f"[fold{fold}] loss": losses.val,
                       f"[fold{fold}] lr": scheduler.get_lr()[0]})

    return losses.avg

def valid_fn(valid_loader, model, criterion, device):

    losses = AverageMeter()
    model.eval()
    preds = []
    start = end = time.time()
    for step, (inputs, labels) in enumerate(valid_loader):
        inputs = collate(inputs)
        for k, v in inputs.items():
            inputs[k] = v.to(device)
        labels = labels.to(device)
        batch_size = labels.size(0)
        with torch.no_grad(): # Torch.no_grad() deactivates autograd engine. Eventually it will reduce the memory usage and speed up computations.
            y_preds = model(inputs)
            loss = criterion(y_preds, labels)
        if CFG.gradient_accumulation_steps > 1:
            loss = loss / CFG.gradient_accumulation_steps
        losses.update(loss.item(), batch_size)
        preds.append(y_preds.to('cpu').numpy())
        end = time.time()
        if step % CFG.print_freq == 0 or step == (len(valid_loader)-1):
            print('EVAL: [{0}/{1}] '
                  'Elapsed {remain:s} '
                  'Loss: {loss.val:.4f}({loss.avg:.4f}) '
                  .format(step, len(valid_loader),
                          loss=losses,
                          remain=timeSince(start, float(step+1)/len(valid_loader))))
    predictions = np.concatenate(preds)
    return losses.avg, predictions
```

# train loop

```python
# ====================================================
# train_loop: train model for each way of splitting fold. 
# return a data Frame that contains original data and predicted data for valid set of this fold.  
# ====================================================
def train_loop(trainDataFrame, foldNumber):
    

    # start training.
    LOGGER.info(f"========== fold: {foldNumber} training ==========")

    # ====================================================
    # loader
    # ====================================================
    
    # get dataFrame of training set of this fold. # Sample that have 'fold' value != foldNumber will used to train model. 
    train_folds = trainDataFrame[trainDataFrame['fold'] != foldNumber].reset_index(drop=True)

    # get dataFrame of validation set of this fold. # Sample that have 'fold' value == foldNumber will used to validate model.
    valid_folds = trainDataFrame[trainDataFrame['fold'] == foldNumber].reset_index(drop=True)

    # valid_labels here is still a dataFrame.
    valid_labels = valid_folds[CFG.target_cols].values
    
    # sent all training data to class DataSet, and get an train object of DataSet class.
    train_dataset = TrainDataset(CFG, train_folds)
    # sent all validation data to class DataSet, and get an valid object of DataSet class.
    valid_dataset = TrainDataset(CFG, valid_folds)

    # used train object created above to create a train_loader object.
    # Data loader used to load data as batch, shuffle data, and sent data to GPU.

    train_loader = DataLoader(train_dataset,
                              batch_size=CFG.batch_size,
                              shuffle=True,
                              num_workers=CFG.num_workers, pin_memory=True, drop_last=True)

    # batch_size when valid is 2 times of batch_size when train.
    valid_loader = DataLoader(valid_dataset,
                              batch_size=CFG.batch_size * 2,
                              shuffle=False,
                              num_workers=CFG.num_workers, pin_memory=True, drop_last=False)

    # ====================================================
    # model & optimizer
    # ====================================================
    model = CustomModel(CFG, config_path=None, pretrained=True)
    torch.save(model.config, OUTPUT_DIR+'config.pth')
    model.to(device)
    
    # return all the positive parameters of model include.
    def get_optimizer_params(model, encoder_lr, decoder_lr, weight_decay=0.0):
        param_optimizer = list(model.named_parameters())
        no_decay = ["bias", "LayerNorm.bias", "LayerNorm.weight"]
        optimizer_parameters = [
            {'params': [p for n, p in model.model.named_parameters() if not any(nd in n for nd in no_decay)],
             'lr': encoder_lr, 'weight_decay': weight_decay},
            {'params': [p for n, p in model.model.named_parameters() if any(nd in n for nd in no_decay)],
             'lr': encoder_lr, 'weight_decay': 0.0},
            {'params': [p for n, p in model.named_parameters() if "model" not in n],
             'lr': decoder_lr, 'weight_decay': 0.0}
        ]
        return optimizer_parameters

    optimizer_parameters = get_optimizer_params(model,
                                                encoder_lr=CFG.encoder_lr, 
                                                decoder_lr=CFG.decoder_lr,
                                                weight_decay=CFG.weight_decay)

    # move optimizer to a AdamW to create a optimizer object.                                                
    optimizer = AdamW(optimizer_parameters, lr=CFG.encoder_lr, eps=CFG.eps, betas=CFG.betas)
    
    # ====================================================
    # scheduler: Used to adjust learning rate theo thoi gian.
    # ====================================================
    def get_scheduler(cfg, optimizer, num_train_steps):
        if cfg.scheduler == 'linear':
            scheduler = get_linear_schedule_with_warmup(
                optimizer, num_warmup_steps=cfg.num_warmup_steps, num_training_steps=num_train_steps
            )
        elif cfg.scheduler == 'cosine':
            scheduler = get_cosine_schedule_with_warmup(
                optimizer, num_warmup_steps=cfg.num_warmup_steps, num_training_steps=num_train_steps, num_cycles=cfg.num_cycles
            )
        return scheduler
    
    
    num_train_steps = int(len(train_folds) / CFG.batch_size * CFG.epochs)
    scheduler = get_scheduler(CFG, optimizer, num_train_steps)

    # ====================================================
    # loop
    # ====================================================

    # Loss. 
    criterion = nn.SmoothL1Loss(reduction='mean') # RMSELoss(reduction="mean")
    
    
    best_score = np.inf

    
    for epoch in range(CFG.epochs):

        start_time = time.time()

        # train
        # train_fn used to train each epoch.
        avg_loss = train_fn(foldNumber, train_loader, model, criterion, optimizer, epoch, scheduler, device)

        # eval
        avg_val_loss, predictions = valid_fn(valid_loader, model, criterion, device)
        
        # scoring
        # score is MCRMSE, score of this comp, scores is a list of MCRMSE of each target.(6 target in this comp)
        score, scores = get_score(valid_labels, predictions)

        elapsed = time.time() - start_time

        LOGGER.info(f'Epoch {epoch+1} - avg_train_loss: {avg_loss:.4f}  avg_val_loss: {avg_val_loss:.4f}  time: {elapsed:.0f}s')
        LOGGER.info(f'Epoch {epoch+1} - Score: {score:.4f}  Scores: {scores}')



        if CFG.wandb:
            # IMPORTANT: LOG the result of epoch to 4 table in wandb. 
            wandb.log({f"[fold{foldNumber}] epoch": epoch+1, 
                       f"[fold{foldNumber}] avg_train_loss": avg_loss, 
                       f"[fold{foldNumber}] avg_val_loss": avg_val_loss,
                       f"[fold{foldNumber}] score": score})
        

        # save model on epoch, if that model have the best score. 
        if best_score > score:
            best_score = score
            LOGGER.info(f'Epoch {epoch+1} - Save Best Score: {best_score:.4f} Model')
            torch.save({'model': model.state_dict(),
                        'predictions': predictions},
                        OUTPUT_DIR+f"{CFG.model.replace('/', '-')}_fold{foldNumber}_best.pth")

    predictions = torch.load(OUTPUT_DIR+f"{CFG.model.replace('/', '-')}_fold{foldNumber}_best.pth", 
                             map_location=torch.device('cpu'))['predictions']
    valid_folds[[f"pred_{c}" for c in CFG.target_cols]] = predictions

    torch.cuda.empty_cache() # free memory of GPU.
    gc.collect() # garbage collection to free memory.
    
    return valid_folds
```

```python
if __name__ == '__main__':
    
    def get_result(oof_df):
        labels = oof_df[CFG.target_cols].values
        preds = oof_df[[f"pred_{c}" for c in CFG.target_cols]].values
        score, scores = get_score(labels, preds)
        LOGGER.info(f'Score: {score:<.4f}  Scores: {scores}')
    
    # CFG.train = true
    if CFG.train:
        
        # create a empty data frame to store all the results of OOF predictions.
        oof_df = pd.DataFrame()
        """
        For each way to split the data
        """
        for fold in range(CFG.n_fold):
            if fold in CFG.trn_fold: 
                # we train a model and store the prediction of valid folds (OOF = Out Of Memory Fold) in the _oof_df
                _oof_df = train_loop(train, fold)
                # append the result of each OOF to a total OOFs.
                oof_df = pd.concat([oof_df, _oof_df])

                LOGGER.info(f"========== fold: {fold} result ==========")
                # get the result of current fold. 
                get_result(_oof_df)
        
        # reset the index to prevent error.
        oof_df = oof_df.reset_index(drop=True)
        LOGGER.info(f"========== CV ==========")
        # get the result of total training process. (all OOF uptill now). Actually, we only score on the valid fold (OOF)
        get_result(oof_df)

        # save the OOF predictions.
        oof_df.to_pickle(OUTPUT_DIR+'oof_df.pkl')
        

        
    if CFG.wandb:
        wandb.finish()

```
